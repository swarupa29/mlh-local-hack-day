```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
import re
```

```python
df=pd.read_csv(r'IMDB-Movie-Data.csv')
```

```python
df.head()
```

## histograms to find the distribution of numeric data

```python
cols=['Year','Rating','Votes','Revenue (Millions)','Metascore','Runtime (Minutes)']
for a in cols:
    plt.hist(df[a],edgecolor='red',bins=10)
    plt.xlabel(a) 
    plt.ylabel("Frequency of occurance")
    plt.show()
```

## Word cloud for fininding the most commonly occuring words

```python
actors=[]
actor=df['Actors']
for g in actor:
    lst=[]
    g=re.sub(r'[^\w]', ' ', g)
    lst=g.split()
    actors=actors+lst
```

```python
unique_actors= " ".join(actors)+" "
  
wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='white', 
                min_font_size = 10).generate(unique_actors)
```

```python
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
print("Most common actors")
plt.show()
```
